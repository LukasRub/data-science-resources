```python
freq = 1/10

t = np.arange(0, 10+freq, freq)
y = np.sin(2*np.pi*freq*t)
z = np.cos(2*np.pi*freq*t)

df = pd.DataFrame(dict(Time=t, Yvals=y, Zvals=z))
display(df.head())

selection = df.iloc[9:15][['Yvals','Zvals']]
display(selection)
```

```python
import os

to_file = 'out.txt'
selection.to_csv(to_file)

if os.path.exists(to_file):
    print(os.path.abspath(to_file))
```
