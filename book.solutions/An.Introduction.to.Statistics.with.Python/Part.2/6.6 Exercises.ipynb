```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from IPython.display import Image

%matplotlib inline
sns.set(style='whitegrid')
```

```python
array = np.arange(0, 10) + 1

mean = np.mean(array)
sample_std = np.std(array, ddof=1)

print('Mean {:.2f}'.format(mean))
print('Sample std {:.2f}'.format(sample_std))
```

```python
mu, sigma = 5, 3
norm = stats.norm(loc=mu, scale=sigma)

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 1000)
y = norm.pdf(x)

plt.figure(figsize=(12,8))
plt.plot(x, y, ls='-', label='PDF of normal distribution (mu=5, sigma=3)')
plt.xticks(np.arange(-2, 12+1))
plt.yticks(np.arange(0.01, 0.15, 0.01))
plt.legend()
plt.show()
```

```python
normal_rvs = norm.rvs(1000)
sem = np.std(normal_rvs, ddof=1) / np.sqrt(len(normal_rvs))

print("{:.3f}".format(sem))
```

```python
plt.figure(figsize=(12,10))
sns.distplot(normal_rvs)
plt.show()
```

```python
significance = 0.05

ci = norm.ppf([significance/2, 1-significance/2])

np.set_printoptions(precision=3)
print(ci)
```

```python
norm = stats.norm()
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x)

plt.figure(figsize=(12, 10))
plt.plot(x,y)
plt.axvline(norm.isf(0.95))
```

```python
plt.figure(figsize=(12, 10))
plt.plot(x,norm.sf(x))
```
