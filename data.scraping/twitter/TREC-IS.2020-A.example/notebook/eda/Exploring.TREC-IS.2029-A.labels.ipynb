```python
labels.eventType.value_counts()
```

Event types need to be gathered from the topics .xml file

```python
topics_path = "../../data/raw/TRECIS-2018-2019.topics.xml"

# Manually adding root node to the xml file
# !sed -i '1s/^/<root>\n/' {topics_path}
# !sed -i '$s/$/\n<\/root>/' {topics_path}
```

```python
import xml.etree.cElementTree as ET

attribs = ['dataset', 'type']

xtree = ET.parse(topics_path)
xroot = xtree.getroot()
rows = []

for node in xroot: 
    res = []
    for el in attribs: 
        if node is not None and node.find(el) is not None:
            res.append(node.find(el).text)
            # Sanity check
            # if el == 'num':
            #     print(node.find('num').text)
        else: 
            res.append(None)
    rows.append(
        {attribs[i]: res[i] for i, _ in enumerate(attribs)}
    )

event_types = pd.DataFrame(rows, columns=attribs)

event_types
```

```python
labels = pd.merge(labels, event_types, left_on='eventID', right_on='dataset')

labels.eventType = labels.type
labels.drop(columns=['type', 'dataset'], inplace=True)

labels.head()
```

### postCategories

```python
_postCategories = labels.postCategories.apply(pd.Series).stack()
_postCategories = pd.get_dummies(_postCategories).sum(level=0)

labels.drop(columns=['postCategories'], inplace=True)
labels = labels.join(_postCategories)

labels[['postID', *_postCategories.columns]].head()
```
