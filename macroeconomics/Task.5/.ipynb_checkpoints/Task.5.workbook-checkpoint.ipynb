```python
#  Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib3
import json


#  Plotting settings
rc = {
    'axes.grid': False,
    'axes.edgecolor': 'lightgrey',
    'axes.spines.top': False,
    'axes.spines.right': False,
    'xtick.color': 'grey',
    'xtick.labelsize': 'small',
    'xtick.bottom': True,
    'xtick.major.size': 2.5,
    'ytick.color': 'grey',
    'ytick.major.size': 2.5,
    'ytick.left': True,
}

sns.set(style='whitegrid', rc=rc)

https = urllib3.PoolManager()
```

The World Bank API example:
1.  `http://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL?format=json`
2.  `http://api.worldbank.org/v2/country/CHL/indicator/DP.DOD.DECD.CR.BC.CD?date=2013Q1:2013Q4`
3.  `http://api.worldbank.org/v2/country/chn;bra/indicator/DPANUSSPB?date=2012M01`

```python
unemployment_indicator = 'SL.UEM.TOTL.ZS'
inflation_indicator = 'FP.CPI.TOTL.ZG'

# For calculating money growth rate
price_level_indicator = 'PA.NUS.PPPC.RF'
gdp_indicator = 'NY.GDP.MKTP.CN'

france_iso3_code = 'FRA'  # See Notes [1] section
year_range = '2000:2019'
```

#### Unemployment data

```python
url = f'https://api.worldbank.org/v2/country/{france_iso3_code}/indicator/{unemployment_indicator}'
url_params = f'?date={year_range}&format=json'

print('Unemployment data source: ', url+url_params)

response = https.request('GET', url+url_params)
response_json = json.loads(response.data)

# print(json.dumps(response_json, indent=2, sort_keys=True))
indicator_explanation = response_json[1][0]['indicator']['value']

france_unemployment_df = pd.DataFrame(response_json[1], columns=['date', 'value']).sort_values('date')
france_unemployment_df.rename(columns=dict(value='unemployment'), inplace=True)
france_unemployment_df['unemployment_explanation'] = indicator_explanation
france_unemployment_df.set_index('date', drop=True, inplace=True)

print(france_unemployment_df.iloc[0,1])

france_unemployment_df
```

#### Inflation data

```python
url = f'https://api.worldbank.org/v2/country/{france_iso3_code}/indicator/{inflation_indicator}'
url_params = f'?date={year_range}&format=json'

print('Inflation data source: ', url+url_params)

response = https.request('GET', url+url_params)
response_json = json.loads(response.data)

# print(json.dumps(response_json, indent=2, sort_keys=True))
indicator_explanation = response_json[1][0]['indicator']['value']

france_inflation_df = pd.DataFrame(response_json[1], columns=['date', 'value']).sort_values('date')
france_inflation_df.rename(columns=dict(value='inflation'), inplace=True)
france_inflation_df['inflation_explanation'] = indicator_explanation
france_inflation_df.set_index('date', drop=True, inplace=True)

print(france_inflation_df.iloc[0,1])

france_inflation_df
```

#### Money growth rate

##### Price level growth

```python
# Getting price level rates
url = f'https://api.worldbank.org/v2/country/{france_iso3_code}/indicator/{price_level_indicator}'
_year_range='1999:2019'
url_params = f'?date={_year_range}&format=json'

print('Price level data source: ', url+url_params)

response = https.request('GET', url+url_params)
response_json = json.loads(response.data)

# print(json.dumps(response_json, indent=2, sort_keys=True))
indicator_explanation = response_json[1][0]['indicator']['value']

france_price_level_df = pd.DataFrame(response_json[1], columns=['date', 'value']).sort_values('date')
france_price_level_df.rename(columns=dict(value='price_level'), inplace=True)
france_price_level_df['price_level_explanation'] = indicator_explanation
france_price_level_df.set_index('date', drop=True, inplace=True)

# Calculating price level growth rate
france_price_level_df['price_level_growth_rate'] = np.nan

for year in france_price_level_df.index[1:]:
    year_ix = france_price_level_df.index.get_loc(year)
    prev_year = france_price_level_df.index[year_ix-1]
    france_price_level_df.loc[year, 'price_level_growth_rate'] = \
        france_price_level_df.loc[year, 'price_level'] / france_price_level_df.loc[prev_year, 'price_level']
    
france_price_level_df['price_level_growth_rate_explanation'] = \
    france_price_level_df['price_level_explanation'] + ' (current year / previous year)'

# Re-ordering columns
cols = france_price_level_df.columns
france_price_level_df = france_price_level_df[[cols[0], cols[2], cols[1], cols[-1]]]

print(france_price_level_df.iloc[0,3])

france_price_level_df
```

##### GDP growth rate

```python
# Getting GDP
url = f'https://api.worldbank.org/v2/country/{france_iso3_code}/indicator/{gdp_indicator}'
_year_range='1999:2019'
url_params = f'?date={_year_range}&format=json'

print('GDP (current LCU) rate data source: ', url+url_params)

response = https.request('GET', url+url_params)
response_json = json.loads(response.data)

# print(json.dumps(response_json, indent=2, sort_keys=True))
indicator_explanation = response_json[1][0]['indicator']['value']

france_gdp_df = pd.DataFrame(response_json[1], columns=['date', 'value']).sort_values('date')
france_gdp_df.rename(columns=dict(value='gdp'), inplace=True)
france_gdp_df['gdp_explanation'] = indicator_explanation
france_gdp_df.set_index('date', drop=True, inplace=True)

# Calculating GDP growth rate
france_gdp_df['gdp_growth_rate'] = np.nan

for year in france_gdp_df.index[1:]:
    year_ix = france_gdp_df.index.get_loc(year)
    prev_year = france_gdp_df.index[year_ix-1]
    france_gdp_df.loc[year, 'gdp_growth_rate'] =  \
        france_gdp_df.loc[year, 'gdp'] / france_gdp_df.loc[prev_year, 'gdp']
    
france_gdp_df['gdp_growth_rate_explanation'] = \
    france_gdp_df['gdp_explanation'] + ' (current year / previous year)'

# Re-ordering columns
cols = france_gdp_df.columns
france_gdp_df = france_gdp_df[[cols[0], cols[2], cols[1], cols[-1]]]

print(france_gdp_df.iloc[0,1])

france_gdp_df
```

##### Calculating money growth rate

```python
france_money_growth_rate_df = pd.DataFrame(index=france_price_level_df.index)
france_money_growth_rate_df['money_growth_rate'] = france_price_level_df['price_level_growth_rate'] + \
                                                   france_gdp_df['gdp_growth_rate']

_explanation = f"{france_price_level_df['price_level_growth_rate_explanation'].values[0]} plus {france_gdp_df['gdp_growth_rate_explanation'].values[0]}"
france_money_growth_rate_df['money_growth_rate_explanation'] = _explanation

print(france_money_growth_rate_df.iloc[0,1])

france_money_growth_rate_df
```

#### Putting it all together

```python
france_df = pd.concat([france_unemployment_df, france_inflation_df, france_price_level_df,
                       france_gdp_df, france_money_growth_rate_df],
                      axis='columns')

# Re-ordering index
france_df.sort_index(inplace=True)

# Re-ordering columns
col_order = ['unemployment', 'inflation', 'price_level', 'price_level_growth_rate', 'gdp', 'gdp_growth_rate',
             'money_growth_rate', 'unemployment_explanation', 'inflation_explanation', 
             'price_level_explanation', 'price_level_growth_rate_explanation', 'gdp_explanation', 
             'gdp_growth_rate_explanation', 'money_growth_rate_explanation']
france_df = france_df.reindex(col_order, axis='columns')

france_df
```

#### Plotting scatterplots

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x='unemployment', y='inflation', hue=france_df.index,
                     data=france_df, 
                     s=50)
plt.title('Unemployment and inflation')
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x='money_growth_rate', y='inflation',
                     data=france_df.iloc[1:-1], 
                     s=50)
plt.title('Money growth rate [2] and inflation')
plt.show()
```

```python
corr_matrix = france_df.corr()
print(f'Correlation between money growth rate and inflation: ', 
      corr_matrix.loc['money_growth_rate', 'inflation'])
```
